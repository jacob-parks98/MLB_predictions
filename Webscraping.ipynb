```python
def get_game(url, game_id):
    game = {'game_id':game_id}
    score_table = soup.find('div', {'class':'scorebox'})
    team_abbr = score_table.find_all('a',{'itemprop':'name'})
    game['away_abbr'] = team_abbr[0]['href'].split('/')[2]
    game['home_abbr'] = team_abbr[1]['href'].split('/')[2]
    date_time = soup.find('div', {'class':'scorebox_meta'}).findAll('div')
    game['date'] = date_time[0].text
    game['time'] = date_time[1].text
    return game

def batting_stats(url,table_number):
    stats_tables = soup.findAll('table', {'class':'stats_table'})
    t = stats_tables[table_number].find('tfoot')
    summary = {x['data-stat']:x.text.strip() for x in t.findAll('td')}
    return summary




```
